```python
import cv2
import os

#    
video_path = '../videos/Overwatch - All Roadhog Skins with All Highlight Intros!.mp4'  #  MP4  
output_dir = 'output_images/Roadhog'  #   

#     
if not os.path.exists(output_dir):
    os.makedirs(output_dir)

#   
cap = cv2.VideoCapture(video_path)

frame_count = 0

while cap.isOpened():
    ret, frame = cap.read()
    if not ret:
        break
   
    frame_filename = os.path.join(output_dir, f'frame_{frame_count:04d}.jpg')
    cv2.imwrite(frame_filename, frame)
    print(f'Frame {frame_count} saved at {frame_filename}')
    
    frame_count += 1

#   
cap.release()
cv2.destroyAllWindows()

print("All frames extracted and saved.")
```
