```python
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt
import seaborn as sns
df = pd.read_csv("Travel details dataset.csv")
df_copy = df.copy()
df
```

```python
plt.figure(figsize=(16,4), dpi=300)
sns.displot(df['Accommodation type'],kde=True, legend=True, aspect=1, color='green')
plt.xticks(rotation=90)
```

```python
plt.figure(figsize=(16,4), dpi=300)
sns.displot(df['Traveler age'],kde=True, legend=True, aspect=1, color='green')
plt.xticks(rotation=90)
```

```python
sns.scatterplot(data=df, x="Duration (days)", y="Accommodation cost", hue='Accommodation type')
```

```python
df['Total cost'] = df['Accommodation cost'] + df['Transportation cost']
df.head()
```

```python
plt.figure(figsize=(16,6), dpi=100)
sns.histplot(data=df.sort_values(by='Total cost', ascending=True), y='Traveler age', x='Traveler nationality', hue='Traveler gender')
plt.xticks(rotation=90)
```

```python
plt.figure(figsize=(10,6), dpi=100)
sns.countplot(data=df, x="Destination")
plt.title("Count of contry visits")
plt.xticks(rotation=90)
```
