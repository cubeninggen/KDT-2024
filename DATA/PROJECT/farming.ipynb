```python
DF1=pd.read_excel('기후변화가 농업생산에 미치는 영향 인식.xlsx')
DF2=pd.read_excel('농업생산_관련_기후변화_대응_방안_20240731143359.xlsx')
DF1
```

```python
labels = DF1.columns
sizes =DF1.loc['40대 이하']

fig, ax = plt.subplots()
ax.pie(sizes,labels=labels)
plt.show()
```
