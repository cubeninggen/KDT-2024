#### 자연어 처리를 위한 전처리
- 정제
    * 토큰화 전후로 진행 됨
    * 100% 제거는 어려움!!
    * 불용어 제거, 빈도에 따른 제거, 길이에 따른 제거

```python
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
```

```python
# NLTK 제공 영어 불용 리스트 가져오기
english_sw=stopwords.words('english')
print(f'영어 불용어 : {len(english_sw)}개')
for idx,_ in enumerate(english_sw[:]):
    print(_,end=' 'if idx%10 else '\n')
```

```python
# NLTK 제공 키릴문자 불용 리스트 가져오기
RU_sw=stopwords.words('Russian')
print(f'러시아어 불용어 : {len(RU_sw)}개')
for idx,_ in enumerate(RU_sw[:]):
    print(_,end=' 'if idx%10 else '\n')
```

```python
# NLTK 제공 키릴문자 불용 리스트 가져오기
ar_sw=stopwords.words('arabic')
print(f'아랍어 불용어 : {len(ar_sw)}개')
for idx,_ in enumerate(ar_sw[:]):
    print(_,end=' 'if idx%10 else '\n')
```

```python
# 토큰에서 불용어 제거하기
text='Завтра будет солнечная погода, и мы пойдем на прогулку по парку.'
```

```python
# 구두점 제거
import string
punct=string.punctuation
for pun in punct:
    text=text.replace(pun,'')
    print(f'punctuation - {pun} : {text}')
```

```python
## 대소문자 통일
text=text.lower()
##  토큰화 진행
tokens=word_tokenize(text)
tokens
```

```python
##불용어 제거
for token in tokens:
    if token in RU_sw:
        #제거
        tokens.remove(token)
```
