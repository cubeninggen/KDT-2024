- [3] 데이터 전처리
    * 기본확인 파악된 네용들 처리
    * 불필요 데이터 삭제 : 0번행 삭제
    * 결측치 데이터 처리 : '전출지별'컬럼 처리
    * 이상치 데이터 처리 : '-'에대한 처리
    * 데이터 자료형 일치 : 형변환

```python
# 3-1 불필요 데이터 처리
copyDF=dataDF.drop(index=0)
copyDF.head(3)
```

```python
# 3-2 결측치 데이터 처리 => isna()/isnull()
copyDF.isna().sum()
```

```python
# - '1984','1985','1986' 컬럼에서 결측치만 추출
a=copyDF[['1984','1985','1986']].isna()
a[a.sum(axis=1)>0]
```

```python
# - '전출지별' 
copyDF['전출지별'].ffill(inplace=True)
```

```python
copyDF[['1984','1985','1986']]=copyDF[['1984','1985','1986']].fillna(0)
copyDF[['1984','1985','1986']]
```

```python
copyDF.isna().sum()
```

```python
# 3-3 이상치 데이터 처리
copyDF.replace('-',0,inplace=True)
```

```python
# 3-4 자료형 변환
d_=dict(zip(copyDF.columns[2:].to_list(),['int']*(copyDF.shape[1]-2)))
copyDF=copyDF.astype(d_)
```

- [4] 데이터 가공
    * 대구에서 전지역 전입 인구수 분석
    * 전출지별 - '대구광역시'
    * 전입지별 - '대구광역시','전국제외'

```python
# - 대구광역시 데이터만 추출
copyDF=copyDF[copyDF['전출지별'] == '대구광역시']
copyDF
```

```python
copyDF.drop(index=55,inplace=True)
```

```python
copyDF.drop(['전출지별'],axis=1,inplace=True)
```

```python
# 2000년도 대구지역에서 타지역으로 이동 인구수 분포
plt.hist(copyDF['2000'],bins=15,histtype='stepfilled')
plt.hist(copyDF['2017'],bins=15,histtype='step')
plt.show()
```

```python
# 2000년도 대구지역에서 타지역으로 이동 인구수 분포
fig=plt.figure(figsize=(10,5))
ax1,ax2=fig.subplots(1,2,sharey=True)
ax1.hist(copyDF['2000'],bins=15,histtype='stepfilled')
ax2.hist(copyDF['2017'],bins=15,histtype='step')
plt.show()
```
