### [산점도 그레프]
- 두개의 컬럼(변수)의 관계를 시각화해서 표현
- 연속형(수치형) 데이터에서 사용

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
F_DATA=r'C:\KDT-2024\PANDAS\DAY_04\auto_mpg.csv'
dataDF=pd.read_csv(F_DATA)
 
```

```python
# 컬럼별 타입확인
dataDF.dtypes
```

```python
# 연비와 관련이 높은 컬럼 즉, 변수확인
# - 상관계수확인
corrSR=dataDF.corr(numeric_only=True)['mpg']
round(corrSR['weight'],2)
corrSR
```

```python
# 시각화
plt.scatter(dataDF.mpg,dataDF.weight,label=f'CORR : {round(corrSR["weight"],2)}')
plt.xlabel('mpg')
plt.ylabel('weight')
plt.legend()
plt.show()
```

```python
# - mpg컬럼과 다름 컬럼관의 관계 시각화
# 그래프 담을 창 생성
fig=plt.figure(figsize=(10,5))
axs=fig.subplots(2,3,sharex=True)
for idx in range(2):
    for col in range(3):
        col_name=corrSR.index[idx*3+col+1]
        axs[idx,col].scatter(dataDF['mpg'],dataDF[col_name],label=f'CORR : {round(corrSR[col_name],2)}')
```
