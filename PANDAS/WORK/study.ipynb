### [인종별로 부모의 학력에 따른 GPA 차이]
- 코카서스계 에서는 부모가 대졸일때 가장 높음
- 아프리카계 에서는 부모가 학사일때 가장 높음
- 아시아계 에서는 None 일때 가장 높음
- 다른 인종에서는 None 일때 가장 높음 

```python
GradeClass={'Caucasian':[CauNGCM,CauHSGCM,CauColGCM,CauBGCM,CauhiGCM],
     'African_American':[AfNGCM,AfHSGCM,AfColGCM,AfBGCM,AfhiGCM],
     'Asian':[AsNGCM,AsHSGCM,AsColGCM,AsBGCM,AshiGCM],
     'Other':[OtNGCM,OtHSGCM,OtColGCM,OtBGCM,OthiGCM]}
GCDF=pd.DataFrame(GradeClass,index=['None','High_School','Some_College',"Bachelor's",'Higher'])
GCDF
```

### [인종별로 부모의 학력에 따른 학점 차이]
- 공통적으로 학사보다 높을때 학점이 높은것을 알수있다

### [결론]
- 인종에 따른 학점의 차이보단 부모가 고학력자이면 학점이 높다는 결과가 나왔다
