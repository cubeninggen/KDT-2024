### [파이 그레프]
- 컬럼(변수)의 데이터 분포 비율 시각화

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
F_DATA=r'C:\KDT-2024\PANDAS\DAY_04\auto_mpg.csv'
dataDF=pd.read_csv(F_DATA)
 
```

```python
# 컬럼별 타입확인
dataDF.dtypes
```

```python
# 특정 컬럼의 데이터 비율
# - 자동차 생산국가에 대한 비율
dataDF['origin'].astype('category')
```

```python
round(dataDF['origin'].value_counts()/dataDF['origin'].count(),2)
```

```python
# - 자동차 생산 국가에 대한 비율 시각화 - bar 그레프
x=dataDF['origin'].value_counts().index.to_list()
h=dataDF['origin'].value_counts().values.tolist()
originBar=plt.bar(x,h,color=['c','m','y'],edgecolor=['k'])
plt.bar_label(originBar,h,label_type='center')
plt.xticks([1,2,3],['US','EU','JP'],rotation=-45)
plt.show()
```

```python
# - 자동차 생산 국가에 대한 비율 시각화 - bar 그레프
plt.pie(h,labels=['US','EU','JP'],autopct='%.2f%%')
plt.show()
```

```python
plt.pie(h,labels=['US','EU','JP'],autopct='%.2f%%',counterclock=False,startangle=90)
plt.show()
```

```python
plt.pie(h,labels=['US','EU','JP'],autopct='%.2f%%',counterclock=False,startangle=-45,explode=[0,0.1,0])
plt.show()
```

```python
plt.pie(h,labels=['US','EU','JP'],autopct='%.2f%%',counterclock=False,startangle=-45,wedgeprops={'linewidth':10,'width':0.6,'edgecolor':'w'},explode=[0,0.2,0])
plt.show()
```
