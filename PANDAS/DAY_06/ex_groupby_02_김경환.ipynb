### [그룹화 처리]
- 주제 : 타이타닉에서 선실 등급에 따른 생존 비율분석

```python
import pandas as pd
import df_util as util
file_name='titanic.csv'
titanicDF=pd.read_csv(file_name)
titanicDF
```

```python
util.checkDF(titanicDF,'titanicDF')
```

```python
serPerClassDF=titanicDF[['survived','class']]
serPerClassDF.isna().sum()
```

```python
groupObj=serPerClassDF.groupby('class')
groupObj
```

```python
groupObj.agg(['count','sum','mean'])
```

- 클레스에 따라서 생존율에 유의미한 치의가 보인다
- 상위 클레스 일수록 생존율이 높은것을 알 수 있다
