### [Matplotlib 활용 데이터 시각화]
- x축과 y축 다양한 처리

```python
from matplotlib import pyplot as plt
```

```python
# 기본
y_data=[1,3,5,7,9]
x_data=['2024-01-01','2024-02-01','2024-03-01','2024-04-01','2024-05-01']
plt.plot(x_data,y_data,marker='o')
plt.show()
```

```python
# x축과 y축 범위설정
y_data=[1,3,5,7,9]
x_data=['2024-01-01','2024-02-01','2024-03-01','2024-04-01','2024-05-01']
plt.plot(x_data,y_data,marker='o')
plt.ylim(bottom=0,top=15)
plt.show()
```

```python
# x축과 y축 눈금설정
y_data=[1,3,5,7,9]
x_data=['2024-01-01','2024-02-01','2024-03-01','2024-04-01','2024-05-01']
plt.plot(x_data,y_data,marker='o')
plt.ylim(bottom=0,top=15)
plt.xticks(rotation=45)
plt.show()
```

```python
# figure(그레프 창) 크기설정
y_data=[1,3,5,7,9]
x_data=['2024-01-01','2024-02-01','2024-03-01','2024-04-01','2024-05-01']
plt.figure(figsize=[10,5]) # 그레프 그리면서 크기설정(순서중요)
plt.plot(x_data,y_data,marker='o')
plt.ylim(bottom=0,top=15)
plt.xticks(rotation=45)
plt.show()
```

```python
# figure(그레프 창) 크기 자동맟춤
y_data=[1,3,5,7,9]
x_data=['2024-01-01','2024-02-01','2024-03-01','2024-04-01','2024-05-01']
plt.plot(x_data,y_data,marker='o')
plt.ylim(bottom=0,top=15)
plt.xticks(rotation=45)
plt.tight_layout()   #모든내용 출력되도록 타이트하게 자동 설정
plt.show()
```

```python
# 눈금(ticks) 설정
y_data=[1,3,5,7,9]
x_data=['2024-01-01','2024-02-01','2024-03-01','2024-04-01','2024-05-01']
plt.plot(y_data,marker='o')
plt.ylim(bottom=-1,top=10)
plt.xlim(left=0,right=5)
font_config={'rotation':45,'fontsize':22}
plt.xticks([1,2,3,4,5],['A','2024-02-01','2024-03-01','2024-04-01','2024-05-01'],**font_config)
plt.tight_layout()   #모든내용 출력되도록 타이트하게 자동 설정
plt.show()
```
