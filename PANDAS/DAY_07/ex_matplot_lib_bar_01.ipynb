### [matplotlib 활용 시각화 - 막대그래프]
- 가장많이 사용됨
- 데이터 비교에 활용

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
fruits = ['apple', 'blueberry', 'cherry', 'orange']
counts = [40, 100, 30, 55]
bar_labels = ['red', 'blue', '_red', 'orange']
bar_colors = ['tab:green', 'tab:blue', 'tab:red', 'tab:orange']
```

```python
# 시각화
plt.barh(fruits,counts,height=0.5,color=bar_colors,label=bar_labels)
plt.legend()
plt.show()
```

```python
plt.bar(fruits,counts,width=0.5,color=bar_colors,label=bar_labels)
plt.plot(counts,'ko--')
plt.legend()
plt.show()
```
