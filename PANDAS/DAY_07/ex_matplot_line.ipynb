### [Matplotlib 활용한 시각화]
- matplotlib.pyplot 모듈을 사용
- 다양한 그레프관련 함수들 & 클레스들 존제

```python
# 패키지명.모듈명
import matplotlib.pyplot as plt
```

```python
# 그레프 그리기 - 선(Line) 그레프
# plot(y축데이터)
plt.plot([1,2,3,4])
plt.xlabel('DATA')
plt.ylabel('RESULT')
plt.show()
```

```python
# 그레프 그리기 - 선(Line)그레프
# plot(X축데이터, Y축데이터)
plt.plot([10,20,30],[5,10,15])
plt.show()
```

```python
# 그레프 그리기 - 선(Line)그레프
# plot(X축데이터, Y축데이터, fmf포멧스트링)
# fmf 포멧스트링 : 그레프 색상, 두께, 스타일 등등 지정
plt.plot([10,20,30],[5,10,15],'r--')
plt.show()
```

```python
# 그레프 그리기 - 선(Line)그레프
# plot(X축데이터, Y축데이터, fmf포멧스트링)
# fmf 포멧스트링 : 그레프 색상, 두께, 스타일 등등 지정
plt.plot([10,20,30],[5,10,15],'mo:')
plt.show()
```

```python
# 그레프 그리기 - 선(Line)그레프
# plot(X축데이터, Y축데이터, line2D속성으로 설정)
# fmf 포멧스트링 : 그레프 색상, 두께, 스타일 등등 지정
# color='cyan', marker='o', linestyle='dashed', markersize=10, linewidth=4
# markeredgecolor => mec
# markeredgewhidth => mew
plt.plot([10,20,30],[5,10,15],color='cyan',marker='o',linestyle='dashed',markersize=10,linewidth=4,mec='m',mew=5)
plt.show()
```

```python
# 그레프 그리기 - 선(Line)그레프
# plt.plot([10,20,30],[25,15,35],'co--')
# plt.plot([10,20,30],[18,29,25],'m>-')
# plt.plot([10,20,30],[22.5,25,29],'y^--')
plt.plot([10,20,30],[25,15,35],'co--',[10,20,30],[18,29,25],'m>-',[10,20,30],[22.5,21,31],'y^--')
plt.show()
```

```python
# 그레프에 부가적인 기능추가
# - 그레프 제목 추가
# - x축, y축 라벨
# - 여러개의 그레프의 경우 볌례 => 그레프의 속성값을 범례에서 사용함
plt.plot([10,20,30],[15,25,20],'co--',label='V') # => plt.legend()
plt.plot([10,20,30],[5,30,28],'mo--',label='2')
plt.title(['MPLIB LINE2D'])
plt.xlabel('Data')
plt.ylabel('Result')
plt.legend() # => label=''
plt.show()
```

```python
# 그레프에 부가적인 기능추가
# - 그레프 제목 추가
# - x축, y축 라벨
# - 여러개의 그레프의 경우 볌례 => 그레프의 속성값을 범례에서 사용함 ,loc 속성으로 위치지정 가능
plt.plot([10,20,30],[15,25,20],'co--',label='V') # => plt.legend()
plt.plot([10,20,30],[5,30,28],'mo--',label='2')
plt.title(['MPLIB LINE2D'])
plt.xlabel('Data')
plt.ylabel('Result')
plt.legend(loc='upper center') # => label=''
plt.show()
```
