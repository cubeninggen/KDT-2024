```python
import pandas as pd
import os
from collections import Counter
TRAIN_PATH='../data/lang/train/'
TEST_PATH='../data/lang/test/'
TRAIN_CSV='../data/lang_train.csv'
TEST_CSV='../data/lang_test.csv'
```

```python
if not os.path.exists(TRAIN_CSV):
    print(f'{TRAIN_CSV}file dose not exists')
else:
    with open(TRAIN_CSV,mode='r')as f:
        alldata=f.read()
        print(f'[File data]{alldata}')
```

```python
#폴더 내부파일 
files=os.listdir(TRAIN_PATH)
for i in files:
    print(i,TRAIN_PATH+i,os.path.exists(TRAIN_PATH+i))
    label=i[:2]
    print(f'[Label] => {label}')
    with open(TRAIN_PATH+i,mode='r')as f:
        alldata=f.read()


    # 대소문자 통일
    alldata=alldata.lower()
    # 알파벳이 아닌 문자제거
    alldata=alldata.replace('\n','')
    print(f'alldata => {len(alldata)}')

    for j in alldata:
        if ord('a')> ord(j) or ord(j) > ord('z'):
            alldata=alldata.replace(j,'')
    print(f'alldata => {len(alldata)}')

    with open(TRAIN_CSV,mode='a')as f:
        cc=dict(Counter(alldata))
        f.write(label+',')
        f.write(','.join([str(value) for value in cc.values()])+'\n')
    
    s_alldata=Counter(alldata)
    print(s_alldata)
    # temp={}
    # for j in alldata:
    #     if j in temp:
    #         temp[j]=temp[j]+1
    #     else:
    #         temp[j]=1
    # print(temp)
```

```python
def make_csv(dir_path,file_path):
    files=os.listdir(dir_path)
    for i in files:
        # 라벨/타겟 추출
        print(i,dir_path+i,os.path.exists(dir_path+i))
        label=i[:2]
        print(f'[Label] => {label}')
        # 데이터 추출
        with open(dir_path+i,mode='r')as f:
            alldata=f.read()


        # 대소문자 통일
        alldata=alldata.lower()
        # 알파벳이 아닌 문자제거
        alldata=alldata.replace('\n','')
        print(f'alldata => {len(alldata)}')

        for j in alldata:
            if ord('a')> ord(j) or ord(j) > ord('z'):
                alldata=alldata.replace(j,'')
        print(f'alldata => {len(alldata)}')

        with open(file_path,mode='a')as f:
            # cc=dict(Counter(alldata))
            a_z=[]
            for _ in range(97,123):
                a_z.append(str((alldata.count(chr(_)))/len(alldata)))
            f.write(label+',')
            f.write(','.join(a_z)+'\n')
        
        s_alldata=Counter(alldata)
        print(s_alldata)

make_csv(TRAIN_PATH,TRAIN_CSV)
make_csv(TEST_PATH,TEST_CSV)
```
